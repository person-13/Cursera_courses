# Загрузка текста(данных) с газеты "Коммерсантъ"

```python
url = "https://www.kommersant.ru/doc/4638951"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

### Данные о заголовке статьи

```python
article_name = soup.find_all('h1', class_='article_name')
article_name = article_name[0].text
article_name
```

### Данные о категории статьи (к какой теме относиться статья)

```python
article_category = soup.find("meta",  {"name":"mywidget:category"}).attrs['content']
article_category
```

##### # Так как в статье может быть больше одной категории то необходимо разбивать по запятым

```python
article_category = article_category.split(",")
```

##### # В данном случае получилось две категории - Экономика, Общество 

### Данные о времени публикации статьи
